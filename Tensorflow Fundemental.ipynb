Constant_tensor

```python
Constant_tensor = tf.constant([[1,2],[3,4]])
```

```python
print(Constant_tensor)
```

Variable_tensor

```python
Variable_tensor = tf.Variable([[5,6],[7,8]])
```

```python
print(Variable_tensor)
```

Row Cancatenate

```python
row_concatenate = tf.concat(values=[Constant_tensor,Variable_tensor],axis=0)
```

```python
print("Row_concatenate:\n",row_concatenate)
```

Column Cancatenate

```python
column_concatenate = tf.concat(values=[Constant_tensor,Variable_tensor],axis=1)
```

```python
print("Column_concatenate:\n",column_concatenate)
```

Zero Tensor

```python
zero = tf.zeros(shape=[2,2],dtype=tf.int32)
```

```python
print("Zero Tensor:\n",zero)
```

One's Tensor

```python
ones = tf.ones(shape=[3,4],dtype=tf.float32)
print("One's Tensor:\n",ones)
```

Tensor Transpose

```python
tensor_transpose = tf.transpose(Constant_tensor)
print("Tensor Transpose:\n",tensor_transpose)
```

Type Conversion

```python
tensor_transpose_float = tf.cast(tensor_transpose,tf.float32)
print("Tensor Transpose_Float:\n",tensor_transpose_float)
```

Tensor Multiplication

```python
tensor_transpose
```

```python
Constant_tensor
```

```python
#NORMAL MULTIPLICATION
multiply_elements = tf.multiply(tensor_transpose,Constant_tensor)
print("multiply_elements:\n",multiply_elements)
```

MATRIX MULTIPLICATION

```python
mat_multiply = tf.matmul(tensor_transpose,Constant_tensor)
print("matrix multiply:\n",mat_multiply)
```

TENSOR DETERMINANT

```python
tensor_transpose_float
```

```python
determinant = tf.linalg.det(tensor_transpose_float)
print(determinant)
```

INVERSE

```python
inverse = tf.linalg.inv(tensor_transpose_float)
print(inverse)
```

IDENTITY MATRIX

```python
identity = tf.eye(num_rows = 3,num_columns = 3,dtype = tf.int32)
print(identity)
```

TENSOR RESHAPE

```python
reshape_con_ten = tf.reshape(tensor = inverse,shape=[1,4])
print(reshape_con_ten)
```
